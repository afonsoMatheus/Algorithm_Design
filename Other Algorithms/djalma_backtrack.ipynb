```python
count = 0

def mvr(present_list, friends):
    
    minor = friends[0]
    
    for i in range(1, len(present_list.keys())):
        
        if(len(present_list[minor]) > len(present_list[friends[i]])):
            minor = friends[i]
            
    return minor

def foward_ver(present_list, friends, itens):
    
    if(itens == []):
        return present_list
        
    for i in range(len(friends)):
         if(itens[len(itens)-1] in present_list[friends[i]]):
            present_list[friends[i]].remove(itens[len(itens)-1])
            
            if(present_list[friends[i]] == []):
                return False
            
    return present_list
    
def djalma_bt(present_list, heur, itens = []):
    
    global count
        
    friends = list(present_list.keys())
    
    if(friends == []):
        return True
    
    if(heur == 'fv'):
        present_list = foward_ver(present_list, friends, itens)
        if(present_list == False):
            return False
    
    if(heur == 'mvr'):
        var = mvr(present_list, friends)
    else:
        var = friends[0]
        
    for i in range(len(present_list[var])):        
                
        count = count + 1
        
        if(present_list[var][i] not in itens):
            
            itens.append(present_list[var][i])
            aux = present_list.pop(var)
            print(aux)

            result = djalma_bt(present_list, heur, itens)
            
            print(present_list)
                                                
            if(result == False):
                present_list[var] = aux
                print(var)
                print(present_list[var])
                itens.remove(present_list[var][i])
            else:
                present_list[var] = itens.pop(len(itens) - 1 )
                return True
    
    return False
    
    
```
