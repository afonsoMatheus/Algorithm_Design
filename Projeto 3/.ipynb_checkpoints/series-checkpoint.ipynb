```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
def distance(v1, v2):
    
    dist = 0
    dist = math.pow(float(v1) - float(v2), 2)

    return dist

def DTW(ts1, ts2):
    
    n = len(ts1) + 1
    m = len(ts2) + 1
    
    mat = [[0 for i in range(m)] for i in range(n)]
    
    for i in range(1, n):
        mat[i][0] = 9999
         
    for i in range(1, m):
        mat[0][i] = 9999
        
    for i in range(1, n):
        for j in range(1, m):
            cost = distance(ts1[i-1], ts2[j-1])
            mat[i][j] = cost + min(mat[i-1][j], mat[i][j-1], mat[i-1][j-1])
    
    return mat[n-1][m-1]
```

```python
df = pd.DataFrame(columns = ["id","x","y"])

ts = [[1,2,1],[1,0,1]]

k = 0
for i in range(len(ts)):
    for j in range(len(ts[i])):
        df.loc[k] = [i] + [j] + [float(ts[i][j])]
        k += 1
                
#df.loc[7] = [2] + [0] + [1.0]
#df.loc[8] = [2] + [0] + [3.0]

#df.loc[9] = [3] + [1] + [2.0]
#df.loc[10] = [3] + [1] + [4.0]

#df.loc[13] = [5] + [1] + [2.0]
#df.loc[14] = [5] + [2] + [4.0]

#df.loc[11] = [4] + [2] + [1.0]
#df.loc[12] = [4] + [3] + [3.0]

#flatui = ["red", "yellow", "black", "black", "black", "black"]

#sns.lineplot(x="x", y="y", data=df, hue = "id", estimator=None, 
             #palette=sns.color_palette(flatui), legend = False)

sns.lineplot(x="x", y="y", data=df, hue = "id", estimator=None, legend = False)
        
print(DTW(ts[0], ts[1]))
    
```
